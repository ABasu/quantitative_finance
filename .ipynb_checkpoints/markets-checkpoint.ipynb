```python
import json, logging, os, sqlite3

import numpy as np
import scipy.signal
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# These are my custom libraries
import alphavantage as av
import timeseries as ts

logging.basicConfig(format = '%(asctime)-25s %(message)s', level = logging.INFO)
```

### Retrieve prices for a ticker from the API

* Retrieve price date from API. 
* Parse the response into a list of (datestring, price) tuples. 
* Create a dataframe with datestring and price columns.

```python
ticker = 'spy'
# av_func = 'TIME_SERIES_WEEKLY_ADJUSTED'
av_func = 'TIME_SERIES_DAILY'

av_response = av.call_api(symbol=ticker, function=av_func, parameters={'outputsize':'full'})
parsed_response = av.parse_response(av_response, function=av_func)
if parsed_response:
    timeseries = ts.timeseries_df(parsed_response) 
    print(timeseries)
```

```python
sns.set_theme(style="darkgrid")
sns.set(rc = {'figure.figsize': (15,5)})
p = sns.lineplot(x="Date", y=ts.rolling_avg(timeseries['Price'], 30, func=np.average), data=timeseries)
p.set_title('{} data for ${}'.format(av_func.replace('_', ' ').title(), ticker.upper()))
p.set_xlim((timeseries['Date'][0], timeseries['Date'][len(timeseries['Date'])-1]))
```

### Up/Down local trends analysis

* Generate new column with percentage move compared to pervious day/week
    * First day is arbitrarily set as 0

```python
pct_move =  [0] + [ts.pct(p, timeseries['Price'][i]) for i, p in enumerate(timeseries['Price'][1:])]
timeseries['Pct_change'] = pct_move
print(timeseries)
```

##### Plot daily fluctuations

```python
sns.set(rc = {'figure.figsize': (15,5)})
p = sns.lineplot(x="Date", y="Pct_change", data=timeseries, linewidth=.5)
p.set_title('{} data for ${}: Daily percentage change'.format(av_func.replace('_', ' ').title(), ticker.upper()))
p.set_xlim((timeseries['Date'][0], timeseries['Date'][len(timeseries['Date'])-1]))
```

##### Plot standard deviations of daily fluctuations over a N day period

* Generate a new column for standard deviations

```python
N = 30

sns.set(rc = {'figure.figsize': (15,5)})
p = sns.lineplot(x="Date", y=ts.rolling_avg(timeseries['Pct_change'], N, func=np.std), data=timeseries, linewidth=1)
p.set_title('{} data for ${}: Std deviation for daily changes over a {} day period'.format(av_func.replace('_', ' ').title(), ticker.upper(), N))
p.set_xlim((timeseries['Date'][0], timeseries['Date'][len(timeseries['Date'])-1]))
p.set_ylim((0, max(timeseries['Pct_change_std'])))
```

### Detect and plot peaks and troughs

```python
peaks = scipy.signal.find_peaks_cwt(timeseries['Price'], 30)
x_peaks, y_peaks = zip(*[(timeseries['Date'][pi], timeseries['Price'][pi]) for pi in peaks])

troughs = scipy.signal.find_peaks_cwt(timeseries['Price']*-1, 30)
x_troughs, y_troughs = zip(*[(timeseries['Date'][pi], timeseries['Price'][pi]) for pi in troughs])


sns.set(rc = {'figure.figsize': (15,10)})
p = sns.lineplot(x="Date", y=ts.rolling_avg(timeseries['Price'], 1, func=np.average), data=timeseries, linewidth=1)
sns.scatterplot(x=x_peaks, y=y_peaks, ax=p, color='green', s=200)
sns.scatterplot(x=x_troughs, y=y_troughs, ax=p, color='red', s=200)
p.set_title('{} data for ${}: Peaks and troughs for {} day window'.format(av_func.replace('_', ' ').title(), ticker.upper(), N))
p.set_xlim((timeseries['Date'][0], timeseries['Date'][len(timeseries['Date'])-1]))
```

### Extract N day slices 

```python
def create_peak_trough_labels(ts, peaks, troughs, N):
    start, end = (int(N/2), int(N/2)) if N % 2 == 1 else (int(N/2), int(N/2-1))

    labels = np.array([0]*len(ts))
    for t in troughs:
        labels[t-start:t+end+1] += 1
    for p in peaks:
        labels[p-start:p+end+1] += 2
    return labels

def get_timeseries_sample(ts, labels, startindex):
    label = labels[startindex]

    i = startindex
    while labels[i]==label and i>=0: i -= 1
    start = i+1
    
    i = startindex
    while labels[i]==label and i<len(ts): i += 1
    end = i
    
    return ts[start:end]
```

```python
N = 6

labels = create_peak_trough_labels(timeseries['Pct_change'], peaks, troughs, N)

t_vecs = [get_timeseries_sample(timeseries['Pct_change'], labels, ti) for ti in troughs]
p_vecs = [get_timeseries_sample(timeseries['Pct_change'], labels, pi) for pi in peaks]

n_vecs = []
while len(n_vecs)<:
    l = 1
    while l != 0:
        i = np.random.randint(len(labels)-1)
        l = labels[i]
    nv = get_timeseries_sample(timeseries['Pct_change'], labels, i)
    if len(nv) >= N:
        n_vecs.append(nv[:N])
```

```python
vectors = n_vecs + t_vecs + p_vecs
vectors = np.stack(vectors)
labels = [0] * len(n_vecs) + [1] * len(t_vecs) + [2] * len(p_vecs) 
print(vectors.shape)

pca = PCA(n_components=2)
transformed = pca.fit_transform(vectors)
x = transformed[:,0]
y = transformed[:,1]

print(transformed.shape)
sns.set(rc = {'figure.figsize': (15,10)})
p = sns.scatterplot(x=transformed[:,0], y=y, hue=labels, s=200)
```

```python
trough_slices = []
last_p = 0
for t in troughs:
    for p in peaks:
        if p > t:
            break
        else:
            last_p = p
    if t - last_p > N:
        last_p = t - N
    trough_slices.append((last_p, t))

peak_slices = []
last_t = 0
for p in peaks:
    for t in troughs:
        if t > p:
            break
        else:
            last_t = t
    if p - last_t > N:
        last_t = p - N
    peak_slices.append((last_t, p))

sns.set(rc = {'figure.figsize': (15,10)})
p = sns.lineplot(x="Date", y="Price", data=timeseries, alpha=.2, linewidth=1)

for ps in peak_slices:
    x = timeseries['Date'][ps[0]:ps[1]]
    y = timeseries['Price'][ps[0]:ps[1]]
    sns.lineplot(x=x, y=y, color='green', linewidth=2, ax=p)
for ts in trough_slices:
    x = timeseries['Date'][ts[0]:ts[1]]
    y = timeseries['Price'][ts[0]:ts[1]]
    sns.lineplot(x=x, y=y, color='red', linewidth=2, ax=p)
p.set_title('{} data for ${}: {} day period leading up to peaks or troughs'.format(av_func.replace('_', ' ').title(), ticker.upper(), N))
p.set_xlim((timeseries['Date'][0], timeseries['Date'][len(timeseries['Date'])-1]))
```

```python
peaks= [[timeseries['Pct_change'][s:e]] for s, e in peak_slices]


fig, (ax0, ax1) = plt.subplots(1,2, figsize=(15,5))
for p in peaks:
    sns.distplot(p, ax=ax0)
```

### Database interface. 

* The first time the database is created, use the following command to set up the tables.

    CREATE TABLE time_series_adjusted (entry integer primary key, ticker text not null, timestamp integer, price real);


```python
db_file = 'market_data.sqlite'

connection = sqlite3.connect(db_file)
cursor = connection.cursor()
```

### Write to database

* Convert timeseries data to (ticker, timestamp, price) tuples.
* Read timeseries for ticker from the database.
* Throw away older values than the most recent timestamp in the database.
* Write the remaining values to the database.

```python
rows = [(ticker, ts.dt64_to_timestamp(t), p) for i, t, p in timeseries.to_records()]
print(rows[:5])
```

```python
cursor.executemany('INSERT INTO time_series_adjusted VALUES(?,?,?);', rows)
```

```python
connection.commit()
```
