```python
import pandas as pd
pd.set_option('display.max_rows', 100)
```

```python
datafile = '../data/michelin_my_maps.csv'
mdf = pd.read_csv(datafile)
mdf = mdf.dropna(subset=['MinPrice','MaxPrice'])
mdf['MaxPrice'] = mdf['MaxPrice'].str.replace(',', '')
mdf['MaxPrice']=[int(m) for m in mdf['MaxPrice'].dropna()]
mdf['MinPrice'] = mdf['MinPrice'].str.replace(',', '')
mdf['MinPrice']=[int(m) for m in mdf['MinPrice'].dropna()]

#mdf['MaxPrice'] = mdf['MaxPrice'].astype('Float64', errors='raise')
# mdf['MinPrice'] = mdf['MinPrice'].str.replace(',', '')
# mdf['MinPrice'] = mdf['MinPrice'].astype('Int64', errors='ignore')
# mdf.dtypes
# mdf.iloc[0]['MinPrice']
```

##### What are the major currencies?

```python
to_usd = {'EUR': 1.11, 'JPY':.0082, 'USD':1, 'GBP':1.31, 'CHF':1.08}
mdf.groupby('Currency', as_index=False, dropna=False).count().sort_values(by='Name', ascending=False)
```

#### What are the major cities?

```python
mdf.groupby('Location', as_index=False, dropna=False).count().sort_values(by='Name', ascending=False)
```

##### Parse out country information from the address field. What are the major countries?

```python
mdf['Country']=[a.split(',')[-1].strip() for a in mdf['Address']]

mdf.groupby('Country', as_index=False, dropna=False).count().sort_values(by='Name', ascending=False)
```

```python
cmdf = mdf[['Country','Name','Location','MinPrice','MaxPrice','Currency','Cuisine','Award']]
usdf = cmdf[cmdf['Country']=='United States']
usdf.groupby('Location', as_index=False).count().sort_values(by='Name', ascending=False)
```

```python
location = 'Chicago'
max = 500
usdf[usdf['Location']==location].groupby('Award').count()
```

```python
usdf[(usdf['Location']==location)&(usdf['MaxPrice']<max)].sort_values(by=['MinPrice', 'MaxPrice'], ascending=True).reset_index(drop=True)
```

```python
gdf = cmdf[cmdf['Country']=='Germany'].sort_values(by=['MinPrice', 'MaxPrice'], ascending=True).reset_index(drop=True)
gdf = gdf[['Name', 'Location', 'MinPrice', 'MaxPrice', 'Cuisine', 'Award']]
gdf.to_csv('bandyts_michelin.csv')
```
